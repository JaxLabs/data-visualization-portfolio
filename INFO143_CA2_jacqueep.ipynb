# The Original: Bar Chart

## Lego Dataset

The original dataset I choice is a summary of the number of lego data sets by lego theme and year. 
Information about this data set can be found through this link [Rebrickable](https://rebrickable.com/downloads/). 


```python
url_lego = 'https://raw.githubusercontent.com/3eltran23/InfoViz/master/LegoTotalSetsPerYear.csv'

df = pd.read_csv(url_lego)
df = df[df['year']> 2017]
df.head(11)
```

## The original plot

```python
labels = ['Star Wars', 'Harry Potter', 'Marvel','Others' ]
qcs =['#1b9e77','#7570b3','#d95f02','#BFBEBE']
width = 0.6 

bottom = np.zeros((len(df),), dtype=int)


fig, ax = plt.subplots(figsize=(15,10))

for pos, label in enumerate(labels):
  
    ax.bar(df['year'], df[label], width=width, label=label, color=qcs[pos], bottom=bottom,edgecolor='white')
    bottom = bottom +  df[label]
 
ax.set_ylabel('# LEGO sets')
ax.legend()
ax.set_title('The Number of LEGO Sets Released by Year', fontsize=26, loc='left')

plt.show()
```

In this original plot a stacked bar chart was bring atenchen to as the title suegests how many lego sets were release in the last 5 years. While the stacked bar chart was a good choice to show over all amounts per year if this plot was just suport to show this data the separation of the franchizes the lego sets belonged to was not needed and the gray coloring of the "Others" catagory was not needed as it need not lend anything to the title.

# Label data Directly

I decided not that the data was about the categories and not the overall amounts it was better to ditch the stacked bar chart for a group bar chart.
I also took the legend and placed it on the top left of the plot spread out and in the same order that the bars appear to aid in looking at the data and not having to glance to the top right to understand what the data was like the original plot.

```python
qual_colors =['#1b9e77','#7570b3','#d95f02','#BFBEBE']

fig, ax = plt.subplots(figsize=(15,7)) 
bar_width = 0.10

x_pos = np.arange(len(df)) #bar location per category

ax.bar(x_pos + 0*bar_width, df["Others"], width=bar_width,  label="Other", color=qual_colors[3])
ax.bar(x_pos + 1*bar_width, df["Marvel"], width=bar_width,  label="Marvel", color=qual_colors[1])
ax.bar(x_pos + 2*bar_width, df["Star Wars"], width=bar_width,  label="Star Wars", color=qual_colors[2])
ax.bar(x_pos + 3*bar_width, df["Harry Potter"], width=bar_width,  label="Harry Potter", color=qual_colors[0])
    
ax.set_ylabel('# LEGO sets', fontsize=14)
ax.set_title('Major Franchises Don\'t Make Up Majority of Lego Sets', fontsize=26, loc='center')
ax.set_xticks(x_pos + bar_width)
ax.set_xticklabels(df['year'], fontsize=14)
ax.legend(loc='upper left',bbox_to_anchor=(-0.01, 1.01), frameon=False, ncol=5, fontsize=14) # here I locate the legend (x=-0.01, y=1.2) upper left. Also I modify the border-color property to red.

ax.tick_params(color=GRAY4)
ax.spines['bottom'].set_color(GRAY4)
ax.spines['left'].set_color(GRAY4)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.show()
```

# Use sufficient contrast

The contrast of the original was fairly good however the gray for the "Others" category was not needed and I wanted to change the colors to colors I find more pleasant while still keeping them distinct and very cohesive for the qualitative data. 

When an image of the plot is passed through [Coblis](https://www.color-blindness.com/coblis-color-blindness-simulator/) the contrast is still very high making this plot acessable to those who are color blind.

```python


qual_colors =['#7fc97f','#386cb0','#fdc086','#beaed4']

fig, ax = plt.subplots(figsize=(15,7)) 
bar_width = 0.10

x_pos = np.arange(len(df))

ax.bar(x_pos + 0*bar_width, df["Others"], width=bar_width,  label="Other", color=qual_colors[3])
ax.bar(x_pos + 1*bar_width, df["Marvel"], width=bar_width,  label="Marvel", color=qual_colors[1])
ax.bar(x_pos + 2*bar_width, df["Star Wars"], width=bar_width,  label="Star Wars", color=qual_colors[2])
ax.bar(x_pos + 3*bar_width, df["Harry Potter"], width=bar_width,  label="Harry Potter", color=qual_colors[0])
    
ax.set_ylabel('# LEGO sets', fontsize=14)

ax.set_title('Major Franchises Don\'t Make Up Majority of Lego Sets', fontsize=26, loc='center')
ax.set_xticks(x_pos + bar_width)
ax.set_xticklabels(df['year'], fontsize=14)
ax.legend(loc='upper left',bbox_to_anchor=(-0.01, 1.01), frameon=False, ncol=5, fontsize=14) # here I locate the legend (x=-0.01, y=1.2) upper left. Also I modify the border-color property to red.

ax.tick_params(color=GRAY4)
ax.spines['bottom'].set_color(GRAY4)
ax.spines['left'].set_color(GRAY4)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.show()
```
