## 2. Designing a static visualization

### 2.1: Choose a plot

```python
sns.kdeplot(data=df, x="People", hue="Year", fill=True, common_norm=False, alpha=0.4)
plt.show()
```

```python
sns.set_theme(style="ticks")

sns.boxplot(x="Sex", y="People", hue="Year", data=df)
```

```python
plt.figure(figsize=(11, 11))
bar1 = sns.barplot(y = df['People'],x = "Year", data=df)
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.barplot(data=df, x="Age", y="People", hue="Year")
```

### 2.2: Add alternative text

A grouped barchart showing volume of population in regads to age groups from 0 to 90 and the years 1900 and 2000. The volume of people in 2000 is significantly lager then in 19000 in all age groups. 

### 2.3: Employ a takeaway title

```python
#takeaway title: Major increase in population from 1900 to 2000

plt.figure(figsize=(15, 10))
sns.barplot(data=df, x="Age", y="People", hue="Year", ci=None)
plt.title('Has there been a population increase from 1900 to 2000?', fontsize=20)
```

## 3. Implement the visualization

```python
sns.color_palette("Set2")
plt.figure(figsize=(14, 10))
sns.barplot(data=df, x="Age", y="People", hue="Year", palette="tab10", ci=None)
plt.legend(title='Year', fontsize=16)
plt.xlabel('Age', fontsize=16);
plt.ylabel('U.S. Population', fontsize=16);
plt.title('Has there been a population increase from 1900 to 2000?', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
```
